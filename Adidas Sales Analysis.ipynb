```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r"C:\Users\–ö–∞—Ä–∏–Ω–∞\Downloads\Adidas_US_Sales_Datasets.csv"

df = pd.read_csv(file_path)

print("‚úÖ –§–∞–π–ª —É—Å–ø—ñ—à–Ω–æ –∑–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–æ!")
print("–†–æ–∑–º—ñ—Ä –¥–∞—Ç–∞—Ñ—Ä–µ–π–º—É:", df.shape)
print("\n–ü–µ—Ä—à—ñ —Å—Ç—Ä–æ–∫–∏:")
display(df.head(5))
```

```python
print("\n–ó–∞–≥–∞–ª—å–Ω–∞ —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—è –ø–æ –¥–∞–Ω–∏—Ö:")
df.info()
```

```python
print("\nüîé –ê–Ω–∞–ª—ñ–∑ –ø—Ä–æ–ø—É—Å–∫—ñ–≤:")

missing_summary = (
    df.isna()
    .sum()
    .reset_index()
    .rename(columns={'index': 'column', 0: 'missing_count'})
)
missing_summary['missing_%'] = (missing_summary['missing_count'] / len(df) * 100).round(2)

display(missing_summary.sort_values('missing_%', ascending=False))
```

```python
duplicate_count = df.duplicated().sum()

print(f"\n–ö—ñ–ª—å–∫—ñ—Å—Ç—å –ø–æ–≤–Ω–∏—Ö –¥—É–±–ª—ñ–∫–∞—Ç—ñ–≤ —Å—Ç—Ä–æ–∫: {duplicate_count}")

if duplicate_count > 0:
    print("\n–ü–µ—Ä—à—ñ 5 –¥—É–±–ª—å–æ–≤–∞–Ω–∏—Ö —Å—Ç—Ä–æ–∫:")
    display(df[df.duplicated()].head())
else:
    print("‚úÖ –î—É–±–ª—ñ–∫–∞—Ç–∏ –≤—ñ–¥—Å—É—Ç–Ω—ñ.")
```
